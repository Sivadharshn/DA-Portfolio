```
data_cleaned = data.copy()
```

**North Star Matric**

```
total_students = len(data_cleaned)
placed_students = data_cleaned["placed"].sum()
not_placed_students = total_students - placed_students
placement_success_rate = placed_students / total_students

print("\nPlacement Summary")
print("Total Students:", total_students)
print("Placed Students:", placed_students)
print("Not Placed Students:", not_placed_students)
print("Placement Success Rate:", round(placement_success_rate * 100, 2), "%")
```

**Key Metrics for Analysis**

```
academic_columns = [
    "cgpa",
    "attendance",
    "tenth_percentage",
    "twelfth_percentage",
]
```

```
technical_skill_columns = [
    "python_skill",
    "c++_skill",
    "java_skill",
    "ml_skill",
    "web_dev_skill",
]
```

```
soft_skill_columns = [
    "communication_skill",
    "teamwork",
    "leadership",
    "problem_solving",
    "time_management",
]
```

```
experience_columns = [
    "internships",
    "projects",
    "github_projects",
    "hackathons",
    "certifications",
]
```

**Average scores for comparison**

```
data_cleaned["academic_score"] = data_cleaned[academic_columns].mean(axis=1)
data_cleaned["technical_skill_score"] = data_cleaned[technical_skill_columns].mean(axis=1)
data_cleaned["soft_skill_score"] = data_cleaned[soft_skill_columns].mean(axis=1)
data_cleaned["experience_score"] = data_cleaned[experience_columns].mean(axis=1)
```

**Placement rate function**

```
def placement_rate_by(column):
    summary = (
        data_cleaned.groupby(column)["placed"]
        .agg(["count", "sum", "mean"])
        .rename(
            columns={
                "count": "total_students",
                "sum": "placed_students",
                "mean": "placement_rate",
            }
        )
        .reset_index()
    )
    summary["placement_rate"] = summary["placement_rate"] * 100
    summary = summary.sort_values("placement_rate", ascending=False)
    return summary
```

**Categorical Analysis**

```
print("\nPlacement Rate by Branch")
print(placement_rate_by("branch"))

print("\nPlacement Rate by College Tier")
print(placement_rate_by("college_tier"))

print("\nPlacement Rate by Gender")
print(placement_rate_by("gender"))

print("\nPlacement Rate by City Tier")
print(placement_rate_by("city_tier"))
```

**Binned Analysis for Important Numerical Columns**

```
data_cleaned["cgpa_band"] = pd.cut(
    data_cleaned["cgpa"],
    bins=[0, 6, 7, 8, 9, 10],
    labels=["Below 6", "6-7", "7-8", "8-9", "9+"],
)

data_cleaned["aptitude_band"] = pd.cut(
    data_cleaned["aptitude_score"],
    bins=[0, 40, 55, 70, 85, 100],
    labels=["Below 40", "40-55", "55-70", "70-85", "85+"],
)

data_cleaned["communication_band"] = pd.cut(
    data_cleaned["communication_skill"],
    bins=[0, 2, 4, 6, 8, 10],
    labels=["0-2", "3-4", "5-6", "7-8", "9-10"],
    include_lowest=True,
)

data_cleaned["backlog_band"] = pd.cut(
    data_cleaned["backlogs"],
    bins=[-1, 0, 1, 2, 10],
    labels=["0", "1", "2", "3+"],
)
```

```
print("\nPlacement Rate by CGPA Band")
print(placement_rate_by("cgpa_band"))

print("\nPlacement Rate by Aptitude Band")
print(placement_rate_by("aptitude_band"))

print("\nPlacement Rate by Communication Band")
print(placement_rate_by("communication_band"))

print("\nPlacement Rate by Backlog Band")
print(placement_rate_by("backlog_band"))
```

**Correlation Analysis**

```
numeric_data = data_cleaned.select_dtypes(include=["int64", "float64"])
correlation_with_placement = (
    numeric_data.corr()["placed"]
    .drop("placed")
    .sort_values(ascending=False)
)

print("\nCorrelation with Placement")
print(correlation_with_placement)
```

**Outlier Check**

```
outlier_columns = [
    "cgpa",
    "attendance",
    "aptitude_score",
    "logical_reasoning",
    "coding_contest_rating",
    "family_income",
]

outlier_summary = []

for column in outlier_columns:
    q1 = data_cleaned[column].quantile(0.25)
    q3 = data_cleaned[column].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    outlier_count = data_cleaned[
        (data_cleaned[column] < lower_limit)
        | (data_cleaned[column] > upper_limit)
    ].shape[0]

    outlier_summary.append(
        {
            "column": column,
            "lower_limit": round(lower_limit, 2),
            "upper_limit": round(upper_limit, 2),
            "outlier_count": outlier_count,
        }
    )

outlier_summary = pd.DataFrame(outlier_summary)
print("\nOutlier Summary")
print(outlier_summary)
```

**Segmentation of Candidates for Business stakeholders**

```
data_cleaned["candidate_segment"] = np.where(
    (data_cleaned["cgpa"] >= 8)
    & (data_cleaned["technical_skill_score"] >= 7)
    & (data_cleaned["aptitude_score"] >= 70)
    & (data_cleaned["backlogs"] <= 1),
    "High Potential",
    np.where(
        (data_cleaned["cgpa"] >= 7)
        & (data_cleaned["technical_skill_score"] >= 5)
        & (data_cleaned["aptitude_score"] >= 55),
        "Moderate Potential",
        "Needs Preparation",
    ),
)

print("\nPlacement Rate by Candidate Segment")
print(placement_rate_by("candidate_segment"))
```

**Visualizations**

```
sns.set_style("whitegrid")

plt.figure(figsize=(6, 4))
sns.countplot(data=data_cleaned, x="placed")
plt.title("Placed vs Not Placed Students")
plt.xlabel("Placement Status")
plt.ylabel("Number of Students")
plt.show()

plt.figure(figsize=(8, 5))
branch_summary = placement_rate_by("branch")
sns.barplot(data=branch_summary, x="branch", y="placement_rate")
plt.title("Placement Rate by Branch")
plt.xlabel("Branch")
plt.ylabel("Placement Rate (%)")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(6, 4))
college_summary = placement_rate_by("college_tier")
sns.barplot(data=college_summary, x="college_tier", y="placement_rate")
plt.title("Placement Rate by College Tier")
plt.xlabel("College Tier")
plt.ylabel("Placement Rate (%)")
plt.show()

plt.figure(figsize=(7, 4))
cgpa_summary = placement_rate_by("cgpa_band")
sns.barplot(data=cgpa_summary, x="cgpa_band", y="placement_rate")
plt.title("Placement Rate by CGPA Band")
plt.xlabel("CGPA Band")
plt.ylabel("Placement Rate (%)")
plt.show()

plt.figure(figsize=(7, 4))
aptitude_summary = placement_rate_by("aptitude_band")
sns.barplot(data=aptitude_summary, x="aptitude_band", y="placement_rate")
plt.title("Placement Rate by Aptitude Band")
plt.xlabel("Aptitude Band")
plt.ylabel("Placement Rate (%)")
plt.show()

plt.figure(figsize=(8, 5))
segment_summary = placement_rate_by("candidate_segment")
sns.barplot(data=segment_summary, x="candidate_segment", y="placement_rate")
plt.title("Placement Rate by Candidate Segment")
plt.xlabel("Candidate Segment")
plt.ylabel("Placement Rate (%)")
plt.xticks(rotation=20)
plt.show()

plt.figure(figsize=(8, 5))
correlation_with_placement.head(10).sort_values().plot(kind="barh")
plt.title("Top Factors Positively Correlated with Placement")
plt.xlabel("Correlation")
plt.ylabel("Factors")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(data=data_cleaned, x="placed", y="cgpa")
plt.title("CGPA Distribution by Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("CGPA")
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(data=data_cleaned, x="placed", y="aptitude_score")
plt.title("Aptitude Score Distribution by Placement Status")
plt.xlabel("Placement Status")
plt.ylabel("Aptitude Score")
plt.show()
```
